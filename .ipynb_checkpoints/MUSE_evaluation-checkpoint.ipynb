<h1>Classification evaluation of MUSE system</h1>



```python
import pandas as pd
import numpy  as np

from utils.constants import DATA_FOLDER
from utils.help_functions import plot_conf_matrix
```

```python
# data_test_1 = pd.read_csv(DATA_FOLDER + 'test_data.csv')
# data_test_1 = data_test_1.rename(columns={"muse": "muse_diagnose"})

data_test_2 = pd.read_csv(DATA_FOLDER + 'super_testovaci.csv')
```

```python
def preprocess(data):
    data_preprocess = data
    data_preprocess['muse_diagnose'] = data_preprocess.muse_diagnose.str.lower().str.replace(" ", "")
    data_preprocess['muse_diagnose'] = data_preprocess.muse_diagnose.str.partition('whencompared')[0]
    return data_preprocess
```

```python
def parse_diagnose(data):
    data_preprocess = preprocess(data)
    valid_labeled_data_muse = data_preprocess.assign(label_muse='unknown')

    valid_labeled_data_muse.loc[valid_labeled_data_muse.muse_diagnose.str.contains('atrialfibrillation'), 'label_muse'] = 'AF'
    valid_labeled_data_muse.loc[valid_labeled_data_muse.muse_diagnose.str.contains('fibrilace'), 'label_muse'] = 'AF'

    valid_labeled_data_muse.loc[valid_labeled_data_muse.muse_diagnose.str.contains('sinusrhythm'), 'label_muse'] = 'SR'
    valid_labeled_data_muse.loc[valid_labeled_data_muse.muse_diagnose.str.contains('sinusový'), 'label_muse'] = 'SR'
    valid_labeled_data_muse.loc[valid_labeled_data_muse.muse_diagnose.str.contains('sinusbrad'), 'label_muse'] = 'SR'
    valid_labeled_data_muse.loc[valid_labeled_data_muse.muse_diagnose.str.contains('sinusovábrad'), 'label_muse'] = 'SR'

    valid_labeled_data_muse.loc[valid_labeled_data_muse.muse_diagnose.str.contains('sinustach'), 'label_muse'] = 'ST'
    valid_labeled_data_muse.loc[valid_labeled_data_muse.muse_diagnose.str.contains('sinusovátach'), 'label_muse'] = 'ST'

    valid_labeled_data_muse.loc[valid_labeled_data_muse.muse_diagnose.str.contains('flutter'), 'label_muse'] = 'FL'
    return valid_labeled_data_muse
```

```python
# Test set 2
parsed = parse_diagnose(data_test_2)
```

```python
af_annotation = parsed[parsed['rhythm'] == "AF"]
print(f"Number of AF cases: {len(af_annotation), round((len(af_annotation)/len(parsed)) * 100, 1)}")

tp = len(af_annotation[af_annotation.label_muse == "AF"])
fn = len(af_annotation[af_annotation.label_muse != "AF"])
tn = len(parsed[(parsed.label_muse != 'AF') & (parsed.rhythm != 'AF')])
fp = len(parsed[(parsed.label_muse == 'AF') & (parsed.rhythm != 'AF')])
```

```python
# metrics
sensitivity = tp / (tp + fn)
print(f'Sensitivity: {round(sensitivity * 100, 2)}')

specificity = tn / (tn + fp)
print(f'Specificity: {round(specificity * 100, 2)}')

precision = tp / (tp + fp)
f1_score = 2 * sensitivity * precision / (sensitivity + precision)

print(f'F1: {round(f1_score, 2)}')
```

```python
# confusion matrix
cm = np.array([[tp, fn], [fp, tn]])
plot_conf_matrix(cm, 'conf_matrix_super_test_muse')
```
