<h2>Plot signal transformation</h2>

```python
def plot_signal_transformation(original, augmented):
    x = np.arange(4096)
    
    plt.plot(x, augmented, '#90AFC5', label='augmentation')
    plt.plot(x, original, '#2A3132', label='original')
    plt.legend(['Baseline Wander Addition', 'Original'])
    
    plt.xlim(1900, 2900)
    plt.ylim(-0.3, 0.8)
    # display
    plt.savefig("augmentace.svg")
    plt.savefig("augmentace.png")
    
    plt.show()
```
